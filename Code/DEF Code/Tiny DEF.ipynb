Generate data:

```python
def generate_popcode_data(ndatapergain, nneuron, mu, sigma, sigtc_sq, gain, noise):
    s = np.random.normal(mu, sigma, ndatapergain)
    sprefs_data = np.tile(sprefs, (ndatapergain, 1))
    r = gain * np.exp(-np.square((np.transpose(np.tile(s, (nneuron, 1))) - sprefs_data))/(2 * sigtc_sq))
    if noise == "poisson":
        r = np.random.poisson(r) + 0.0
    #questionable? does throwing out all 0 trials screw up training?
    r = r[~(np.sum(r, axis=1)==0)]
    s = s[~(np.sum(r, axis=1)==0)]
    ar = np.sum(r, axis=1)/sigtc_sq
    br = np.sum(r*sprefs_data[1:(len(r)+1)], axis=1)/sigtc_sq
    #without correction:
    #br = np.sum(r*sprefs_data, axis=1)/sigtc_sq
    mean = br/ar
    var = 1/ar
    return r, s, mean, var
r, actual, mean, var = generate_popcode_data(ndatapergain, nneuron, mu, sigma, sigtc_sq, 1, "poisson")
print r
print actual
print mean
print var
```

Some useful functions:

```python
def relu(zTw):
    return max(min_val, zTw)

def g_norm(z, loc, scale):
    d_loc = (z - loc)/scale**2
    d_scale = -(1/scale) + ((z - loc)**2)/(scale**3)
    return [d_loc, d_scale]

def kl(mu1, mu2, sig1, sig2):
    return np.log(sig2/sig1) + (sig1**2 + (mu1 - mu2)**2)/(2 * sig2**2) - .5
```
